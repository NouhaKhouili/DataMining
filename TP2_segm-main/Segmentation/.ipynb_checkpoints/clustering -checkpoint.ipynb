# <p style="color:#e320c2" >Introduction: </p>
Le clustering est une mthode d'analyse statistique utilise pour organiser des donnes brutes en silos homognes. 

A l'intrieur de chaque grappe, les donnes sont regroupes selon une caractristique commune. 

L'outil d'ordonnancement est un algorithme qui mesure la proximit entre chaque lment  partir de critres dfinis.
Exemple de cluster : kameans 


```python

import numpy as np
import cv2
from sklearn.cluster import KMeans
from sklearn.cluster import MiniBatchKMeans
import skimage.io
import matplotlib.pyplot as plt 
import matplotlib.image as mpimg 

image = mpimg.imread('cat.jpg')
imgplot = plt.imshow(image)
plt.show()

(h1, w1) = image.shape[:2]

image = cv2.cvtColor(image, cv2.COLOR_BGR2LAB)

image = image.reshape((image.shape[0] * image.shape[1], 3))

clt = KMeans(n_clusters = 5)

labels = clt.fit_predict(image)
quant = clt.cluster_centers_.astype("uint8")[labels]

#reshape the feature vectors to images
quant = quant.reshape((h1, w1, 3))
image  = image.reshape((h1, w1, 3))

# convert from L*a*b* to RGB
quant = cv2.cvtColor(quant, cv2.COLOR_LAB2BGR)
image  = cv2.cvtColor(image, cv2.COLOR_LAB2BGR)

cv2.imwrite('cat.png', quant)


cv2.imread('cat.jpg')
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# <p style="color:#028A93">Partie bruit d'image:</p>
A chaque tape de l'acquisition d'une scne, des perturbations (rayures, poussires, camra, amplification, quantification) vont dtriorer la "qualit" de l'image. Ces perturbations sont regroupes sous le nom de "bruit d'image".
quelques exemples : 
<p>"gaussian", "localvar","poisson", "salt", "pepper","s&p", "speckle"</p>




```python
import skimage.io
import matplotlib.pyplot as plt
img_path="cat.jpg"
img = skimage.io.imread(img_path)/255.0

def plotnoise(img, mode, r, c, i):
    plt.subplot(r,c,i)
    if mode is not None:
        gimg = skimage.util.random_noise(img, mode=mode)
        plt.imshow(gimg)
    else:
        plt.imshow(img)
    plt.title(mode)
    plt.axis("off")

plt.figure(figsize=(18,24))
r=4
c=2
plotnoise(img, "gaussian", r,c,1)
plotnoise(img, "localvar", r,c,2)
plotnoise(img, "poisson", r,c,3)
plotnoise(img, "salt", r,c,4)
plotnoise(img, "pepper", r,c,5)
plotnoise(img, "s&p", r,c,6)
plotnoise(img, "speckle", r,c,7)
plotnoise(img, None, r,c,8)
plt.show()
```

# <p style="color:#028A93">Partie bruitage et filtrage de l'image:</p>
dans cette on applique un type de bruit sur l'image cat puis on applique le filtrage 

```python
from skimage.restoration import denoise_tv_chambolle
import skimage.io
import matplotlib.pyplot as plt
# Apply total variation filter denoising
img_path="cat.jpg"
img = skimage.io.imread(img_path)/255.0


def plotnoise(img, mode, r, c, i):
    plt.subplot(r,c,i)
    if mode is not None:
        gimg = skimage.util.random_noise(img, mode=mode)
        plt.imshow(gimg)
    else:
        plt.imshow(img)
    plt.title(mode)
    plt.axis("off")



plt.figure(figsize=(20,24))
r=4
c=2
plotnoise(img, "s&p", r,c,3)
plt.show()
denoised_image = denoise_tv_chambolle(img,weight=0.1,multichannel=True)

plt.imshow(denoised_image)
```
